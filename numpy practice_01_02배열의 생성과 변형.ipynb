```python
import numpy as np
```

```python
x = np.array([1, 2, 3])
print(x.dtype)
```

```python
x = np.array([1.0, 2.0, 3.0])
x.dtype
```

정수와 실수가 섞인 배열
dtype을 명시하지 않았지마느 실수가 하나라도 있기 때문에 넘파이는 해당 배열의 dtype을 float64로 추측(data가 손실되지 않는 방향으로 자동 변환)

```python
x = np.array([1, 2, 3.0])
x.dtype
```

```python
x = np.array([1, 2, 3], dtype = 'f')
x.dtype
```

```python
x[0] + x[1]
```
