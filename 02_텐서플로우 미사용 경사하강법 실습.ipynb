```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#  x  y  
data = [[2, 91], [4, 93], [6, 91], [8, 97]]
x = [i[0] for i in data]
y = [i[1] for i in data]
```

```python
# 
plt.figure(figsize=(8, 5))
plt.scatter(x, y)
plt.show()
```

```python
#   x y   (       )

x_data = np.array(x)
y_data = np.array(y)

# a  b  
a = 0
b = 0
```

```python
# 
lr = 0.05

#   (0    +1)
epochs = 2001

# 
for i in range(epochs): #  
    y_pred = a * x_data + b #y  
    error = y_data - y_pred #  
    #  a  
    a_diff = -(1/len(x_data)) * sum(x_data * (error))
    #  b  
    b_diff = -(1/len(x_data)) * sum(y_data - y_pred)
    a = a - lr * a_diff #    a 
    b = b - lr * b_diff #   b 
    if i % 100 == 0: #100    a, b 
        print("epoch=%.f,=%.04f, =%.04f" % (i, a, b))
```

```python
#       
y_pred = a * x_data + b
plt.scatter(x, y)
plt.plot([min(x_data), max(x_data)], [min(y_pred), max(y_pred)])
plt.show
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = [[1, 0.2], [2, 0.3], [3, 0.5], [4, 0.6], [5, 0.9], [6, 0.95], [7, 1.1], [8, 1.5]]
x = [i[0] for i in data]
y = [i[1] for i in data]
plt.figure(figsize=(8,5))
plt.scatter(x, y)
plt.show()
x_data = np.array(x)
y_data = np.array(y)
a = 0
b = 0
lr = 0.03
epochs = 1001
for i in range(epochs):
    y_pred = a * x_data + b
    error = y_data - y_pred
    a_diff = 2/len(x_data) * sum((a * x_data + b - y_data) * x_data)
    b_diff = 2/len(x_data) * sum(a * x_data + b - y_data)
    a = a - lr * a_diff
    b = b - lr * b_diff
    if i % 10 == 0:
        print("epoch=%.f, =%.04f, =%.04f" % (i, a, b))
y_pred = a * x_data + b
plt.scatter(x, y)
plt.plot([min(x_data), max(x_data)], [min(y_pred), max(y_pred)])
plt.show()
```
