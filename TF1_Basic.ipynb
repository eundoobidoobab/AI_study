```python
import tensorflow as tf

a = tf.constant(7.0, name = 'data1')
b = tf.constant(3.0, name = 'data2')
c = tf.constant(2.0, name = 'data3')

v = a * b / c

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    print(sess.run(v))
```
